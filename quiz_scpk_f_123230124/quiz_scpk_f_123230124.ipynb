## Untuk semua import dari library luar taruh disini.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## No 1. Operasi Matriks 1

```python
A = np.array([[8,12],[5,9]])
B = np.array([[7,9],[5,3]])

print (f"hasil perkalian matriks A dengan matriks B\n {np.dot(A,B)}")
```

## No 2. Operasi Matriks 2

```python
A = np.array([[5,4],[2,-1]])
B = np.array([68.8,8.8])

A_inv = np.linalg.inv(A)
x = np.dot(A_inv, B)

print(f"jadi nilai [x y] adalah \n {x}")
```

## No 3. Data Preprocessing (Import)

```python
data = pd.read_csv("rekap_bulanan.csv")
print(data)
```

## No 4. Data Preprocessing (Informasi)

```python
print(data.info())
```

## No 5. Data Preprocessing (Cek Data NaN)

```python
data_null = data.isnull()
row_null = data[data_null.any(axis=1)]
print (row_null)
```

## No 6. Data Preprocessing (Hapus Data NaN)

```python
data1 = data
data1 = data1.dropna()
print(data1)
```

## No 7. Data Preprocessing (Simpan Data)

```python
data1.to_csv('rekap_cleaned.csv', index=False)
```

## No 8. Data Visualization (Python GUI) 1

```python
bulan = data1['Bulan'].to_numpy()
pengeluaran = data1['Pengeluaran'].to_numpy()

plt.figure(figsize=(10,5))
plt.plot(bulan, pengeluaran, marker = 'o', linestyle='-')

plt.xlabel("bulan")
plt.ylabel("pengeluaran")

plt.show
```

## No 9. Data Visualization (Python GUI) 2

```python
bulan = data1['Bulan'].to_numpy()
pemasukan = data1['Pemasukan'].to_numpy()

plt.figure(figsize=(10,5))
plt.plot(bulan, pemasukan, marker = 'o', linestyle='-')

plt.xlabel("bulan")
plt.ylabel("pemasukan")

plt.show
```

## No 10. Data Visualization (Python GUI) 3

```python


plt.figure(figsize=(10,5))
plt.plot(bulan, pemasukan, label="pemasukan" ,marker = '*', linestyle ='--')
plt.plot(bulan, pengeluaran, label="pengeluaran", marker = '*', linestyle =':')

plt.xlabel("Bulan")
plt.legend()

plt.show
```
