### 3. Grafik dengan python

3.1 Import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

3.2 Grafik sederhana

```python
x = [1,2,3,4]
y = [1,4,8,16]

plt.ylabel("sb y")
plt.xlabel("sb x")
plt.title("Grafik nih bro")
plt.plot(x,y)
plt.show()
```

3.3 Modifikasi grafik

```python
x = np.array([1,2,3,4])
y = x**2

plt.ylabel("sb y",
           fontsize=10,
           color = 'blue',
           labelpad = 25)
plt.xlabel("sb x",
           fontsize=10,
           color = 'blue',
           labelpad = 25)
plt.title("Grafik nih bro",
          fontweight = 'bold',
          fontsize = 13,
          fontstyle = 'italic',
          # font name = 'Arial',
          color = 'blue',
          pad = 20
          )
plt.plot(x,y)
plt.show()
```

3.4 Scatter

```python
x = np.array([1,2,3,4])
y = x**2

plt.ylabel("sb y",
           fontsize=10,
           color = 'blue',
           labelpad = 25)
plt.xlabel("sb x",
           fontsize=10,
           color = 'blue',
           labelpad = 25)
plt.title("Grafik nih bro",
          fontweight = 'bold',
          fontsize = 13,
          fontstyle = 'italic',
          # font name = 'Arial',
          color = 'blue',
          pad = 20
          )
plt.plot(x,y,
        #  color = '#82C8E5',
        #  marker= 'h',
        #  linestyle = ':',
         'b-.^',
         linewidth=2,
         markersize=10
         )
plt.show()
```

3.5 Grid

```python
x = np.array([1,2,3,4])
y = x**2

plt.ylabel("sumbu y",
           fontsize=13,
           color = 'blue',
           labelpad = 25)
plt.xlabel("sumbu x",
           fontsize=13,
           color = 'blue',
           labelpad = 25)
plt.title("Grafik nih bro",
          fontweight = 'bold',
          fontsize = 13,
          fontstyle = 'italic',
          # font name = 'Arial',
          color = 'blue',
          pad = 20
          )
plt.plot(x,y,
        #  color = '#82C8E5',
        #  marker= 'h',
        #  linestyle = ':',
         'b-.^',
         linewidth=2,
         markersize=10
         )
plt.grid(True,
         linestyle=':',
         color = 'c',
         alpha = 0.5,
         linewidth = 1)
plt.show()
```

3.6 Legenda

```python
y2 = x**3

plt.ylabel("sumbu y",
           fontsize=10,
           color = 'black',
           fontweight = 'bold',
           labelpad = 25)
plt.xlabel("sumbu x",
           fontsize=10,
           color = 'blue',
           fontweight = 'bold',
           labelpad = 25)
plt.title("Grafik nih bro",
          fontweight = 'medium',
          fontsize = 10,
          fontstyle = 'italic',
          # font name = 'Arial',
          color = 'blue',
          pad = 20
          )
plt.plot(x,y2,
        #  color = '#82C8E5',
        #  marker= 'h',
        #  linestyle = ':',
         'k-.^',
         linewidth=1,
         markersize=8,
         label = 'Y = 2X'
         )
plt.plot(x,y,
        #  color = '#82C8E5',
        #  marker= 'h',
        #  linestyle = ':',
         'b-.^',
         linewidth=1,
         markersize=8,
         label = 'Y = 2X'
         )
plt.grid(True,
         linestyle=':',
         color = 'b',
         alpha = 0.5,
         linewidth = 1)
plt.legend()
plt.show()
```

```python
# X = np.arange(10)
# X = np.arange(1,10)
# X = np.arange(1,10,2)
# X = {-10,10 | X = R}
x = np.arange(-1,1.01,0.1)
y = np.pow(x,2)
y2 = np.square(x)

plt.plot(x,-y, label = 'Y = X^2')
plt.plot(x,y2, label =' Y = akar X')
plt.legend(loc='upper center')
```
