### Matriks

```python
import numpy as np
```

 Diketahui suatu matriks A = 8 12
5 9
dan matriks B = 7 9
5 3
. Buatkan
hasil dari perkalian antara matriks A dengan matriks B!


```python
A = np.array([[8,12],[5,9]])
B = np.array([[7,9],[5,3]])

hasil = np.dot(A,B)
print(hasil)
```

Diketahui suatu persamaan sebagai berikut :
P1 => 5x + 4y = 68.8
P2 => 2x - y = 8.8
Menggunakan Python Matriks, berikan nilai x & y!

```python
A = np.array([[5,4],[2,-1]])
B = np.array([68.8,8.8])

hasil = np.linalg.solve(A,B)
print(hasil)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hai = pd.read_csv("rekap_bulanan.csv")
print(hai)
```

Tampilkan informasi dari data .csv yang sudah dibaca sebelumnya

```python
print(hai.info())
```

Tampilkan lagi semua datanya. Namun, yang keluar harus data yang

Buatlah supaya semua baris yang memiliki data NaN akan terhapus

```python
print(hai.dropna())
```

Simpan data yang sudah tidak ada nilai NaN nya kedalam file .csv baru
dengan nama rekap_cleaned.csv.

```python
hai.dropna().to_csv('data_bersih.csv')
```

Ambil data pada rekap_cleaned.csv, buat grafik antara data Bulan
(Sumbu X) dengan data Pengeluaran (Sumbu Y) *styling dibebaskan.

```python
hai2 = pd.read_csv('data_bersih.csv')

X = hai2['Bulan']
Y = hai2['Pengeluaran']

plt.plot(X,Y, marker='o')
plt.ylabel('Pengeluaran')
plt.xlabel('Bulan')
```

Ambil data pada rekap_cleaned.csv, buat grafik perbandingan antara
Pemasukan dengan Pengeluaran (Sumbu Y) berdasarkan data Bulan (Sumbu X)
*styling dibebaskan.

```python
Y2 = hai2['Pemasukan']

plt.plot(X,Y2, label='Pemasukan', marker='o')
plt.plot(X,Y, label='Pengeluaran',marker='o')
plt.legend(title='Informasi')
```
